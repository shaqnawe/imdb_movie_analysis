```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb-movies.csv')
df.isnull().sum()
```

### Drop columns without neccesary information and remove all records with no financial information -- Pay close attention to things that don't tell you anything regarding financial data

```python
df_clean = pd.read_csv('imdb-movies.csv')
```

```python
df_clean.dropna(inplace=True)
df_clean.isnull().sum()
df_clean.count()
```

```python
drop_tables = ['popularity','cast','homepage','tagline','overview','keywords','vote_count','vote_average','imdb_id','runtime']

for column in drop_tables:
    df_clean.drop([column], axis=1,inplace=True)
```

```python
df_clean.head(2)
```

### Data Cleaning

```python
# Delete all records with null, or empty values

```

#### Here's a helpful hint from my own analysis when I ran this the first time. This may help shed light on what your data set should look like.

#### If I created one record for each the `production_companies` a movie was release under and one record each for `genres`<br>and tried to run calculations, it wouldn't work because for many records, the amount of `production_companies`<br>and `genres` aren't the same, so I'll create 2 dataframes; one w/o a `production_companies` column and one w/o a `genres` columns

```python
df_production_companies = df_clean.copy()
df_production_companies.drop(['genres'],axis=1,inplace=True)
df_production_companies.head(2)
# df_production_companies.count()
```

```python
df_genres = df_clean.copy()
df_genres.drop(['production_companies'],axis=1,inplace=True)
df_genres.head(2)
# df_genres.count()
```

```python
# prod_comp_split = df_production_companies[df_production_companies.production_companies.str.contains('/')]
# prod_comp_split.head(2)
```

```python
h1=df_production_companies.copy()
h1['production_companies'] = h1['production_companies'].apply(lambda r:r.split('|'))
h1 = h1.explode('production_companies',ignore_index=True)
h1.head()
```

```python
h2=df_genres.copy()
h2['genres'] = h2['genres'].apply(lambda r:r.split('|'))
h2 = h2.explode('genres',ignore_index=True)
h2.head()
```

```python
# h2.info()
# h2.isnull().sum()
h2.to_csv('genres_df.csv')
```

```python
# h1.info()
# h1.isnull().sum()
h1.to_csv('pc_df.csv')
```

## Exploratory Data Analysis

> Use Matplotlib to display your data analysis

```python
movie_count = h1.groupby('production_companies').id.count().nlargest()
movie_count
```

### Which production companies released the most movies in the last 10 years? Display the top 5 production companies.

```python
last_ten = h1.groupby('production_companies').count()
last_ten.query('release_year >= 2012 and release_year <= 2022')
l10_ans = last_ten.nlargest(5, 'original_title')
l10_ans

l10_ans['director'].plot(legend=False, kind='bar', grid=True, color=['cyan'], figsize=(15,10))
plt.title('Movies released between 2012 and 2022')
plt.xlabel('Production Companies', fontsize=10)
plt.ylabel('Number of Movies', fontsize=10);
```

### What 5 movie genres grossed the highest all-time?

```python
mg = pd.read_csv('genres_df.csv')
top_mg = mg.groupby('genres').genres.count().nlargest()
top_mg
```

```python
top_mg.plot(kind='pie', y='genres', subplots=True, autopct='%.2f', figsize=(10,10));
# plt.title('Highest grossing genres')
# plt.xlabel('Genres', fontsize=10)
# plt.ylabel('Number of Movies', fontsize=10);
```

### Who are the top 5 grossing directors?

```python
top_dir = mg.groupby('director').revenue.sum().nlargest()
top_dir
```

```python
top_dir.plot(kind='bar', grid=True, color=['red'], figsize=(15,10))
plt.title('Top 5 grossing Directors')
plt.xlabel('Directors', fontsize=10)
plt.ylabel('Revenue (billions)', fontsize=10);
```

### Compare the revenue of the highest grossing movies of all time.

```python
revenue = mg.groupby('original_title').revenue.sum().nlargest(5)
revenue
```

```python
revenue.plot(kind='bar', grid=True, color=['orange'], figsize=(15,10))
plt.title('Highest Grossing movies')
plt.xlabel('Movie Name', fontsize=10)
plt.ylabel('Revenue (billions)', fontsize=10);
```
